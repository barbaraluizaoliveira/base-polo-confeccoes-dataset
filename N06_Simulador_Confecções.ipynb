```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_confeccoes = pd.read_csv("/content/base_polo_confeccoes.xlsx - Base de dados.csv")
df_confeccoes.head()
```

#### Questão 01 : Uma empresa do setor têxtil está analisando os dados de exportação de roupas por país e gênero. Com base na tabela abaixo, deseja-se identificar o país com a maior proporção de exportações femininas em relação ao total de exportações (feminino + masculino).

Pergunta:
Qual país possui a maior proporção de exportações femininas em relação ao total exportado?

```python
df_grouped = df_confeccoes.groupby(["País de Exportação", "Gênero do Consumidor"])["Receita Total (R$)"].sum().reset_index()
df_grouped.head()
```

```python
# Antes de pivotar: você tem dados em formato longo (linhas com "País", "Gênero", "Receita").
# Após pivotar: você transforma em formato largo, com uma linha por país e uma coluna para cada gênero.
#Isso facilita análises comparativas e cálculos (somar, subtrair, calcular proporção
# resumo:
#Pivotar = reorganizar a tabela para transformar valores de uma coluna (ex: “Gênero”) em colunas separadas, agrupando por outra coluna (ex: “País”

df_pivot = df_grouped.pivot(index="País de Exportação", columns="Gênero do Consumidor", values="Receita Total (R$)").fillna(0)
df_pivot.head()
```

```python
# Padroniza os valores da coluna "Gênero do Consumidor" (remove espaços e capitaliza)
df_confeccoes['Gênero do Consumidor'] = (
    df_confeccoes['Gênero do Consumidor']
    .str.strip()
    .str.capitalize()
)

# Converte os valores da coluna "Receita Total (R$)" para float (remove R$, milhar e troca vírgula por ponto)
df_confeccoes['Receita Total (R$)'] = (
    df_confeccoes['Receita Total (R$)']
    .str.replace('R$', '', regex=False)
    .str.replace('.', '', regex=False)
    .str.replace(',', '.', regex=False)
    .str.strip()
    .astype(float)
)

# Agrupa os dados por país e gênero, somando as receitas
df_grouped = df_confeccoes.groupby(
    ["País de Exportação", "Gênero do Consumidor"]
)["Receita Total (R$)"].sum().reset_index()

# Reorganiza os dados em formato de tabela com gêneros como colunas
df_pivot = df_grouped.pivot(
    index="País de Exportação",
    columns="Gênero do Consumidor",
    values="Receita Total (R$)"
)

# Remove o nome do índice das colunas (apenas estético)
df_pivot.columns.name = None

# Calcula a diferença de receita entre Feminino e Masculino
df_pivot["Diferença"] = df_pivot["Feminino"] - df_pivot["Masculino"]

# Exibe o resultado
df_pivot.head()

```

```python
# Calcula total exportado e proporção feminina
df_pivot["Total"] = df_pivot["Feminino"] + df_pivot["Masculino"]
df_pivot["Proporção_Feminino"] = df_pivot["Feminino"] / df_pivot["Total"]
df_pivot["Proporção (%)"] = (df_pivot["Proporção_Feminino"] * 100).round(1)

# Exibe a tabela com percentuais
print(df_pivot[["Feminino", "Masculino", "Total", "Proporção (%)"]])

# Descobre o país com maior proporção feminina
pais_maior = df_pivot["Proporção_Feminino"].idxmax()
maior_valor = df_pivot["Proporção_Feminino"].max()

```

#### A empresa deseja identificar o canal de vendas com maior estabilidade de faturamento entre 2022 e 2024. Isso ajudará na tomada de decisão para manter previsibilidade nas receitas.

Pergunta:
Qual canal apresentou menor variação de faturamento entre os anos de 2022 a 2024, indicando maior estabilidade nos resultados?

```python
#ver colunas da tabela
df_confeccoes.columns
#ver quais canais de vendas existem na coluna "Canal de Venda"
df_confeccoes["Canal de Venda"].unique()
```

```python
df_confec_copy = df_confeccoes.copy()
#remover $ dos valores de "receita Total"
df_confec_copy['Receita Total (R$)'] = df_confec_copy['Receita Total (R$)'].str.replace('R$', '', regex=False)
df_confec_copy['Receita Total (R$)'] = df_confec_copy['Receita Total (R$)'].str.strip()
df_confec_copy['Receita Total (R$)'] = df_confec_copy['Receita Total (R$)'].str.replace('.', '', regex=False)
df_confec_copy['Receita Total (R$)'] = df_confec_copy['Receita Total (R$)'].str.replace(',', '.', regex=False)
df_confec_copy['Receita Total (R$)'] = df_confec_copy['Receita Total (R$)'].astype(float)
df_confec_copy.head()
```

```python
#ver o tipo da coluna receita total
df_confec_copy['Receita Total (R$)'].dtype
#somar valores de receita total pelo canal de vendas
df_confec_copy.groupby(["Canal de Venda"])["Receita Total (R$)"].sum()
```

```python
df_anual = df_confec_copy.groupby(['Ano', 'Canal de Venda'])['Receita Total (R$)'].sum().reset_index()
df_anual.head()
#Pivotar para que cada canal fique numa linha e os anos como colunas:
df_pivot = df_anual.pivot(index='Canal de Venda', columns='Ano', values='Receita Total (R$)')
df_pivot.head()
#verificar variação, incluindo uma coluna "variação" para analisar melhor
df_pivot['Variação'] = df_pivot.max(axis=1) - df_pivot.min(axis=1)
canal_mais_estavel = df_pivot.sort_values(by='Variação').head(3)
print( canal_mais_estavel)
```
