```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_confeccoes = pd.read_csv("/content/base_polo_confeccoes.xlsx - Base de dados.csv")
df_confeccoes.head()
```

#### Questão 01 : Uma empresa do setor têxtil está analisando os dados de exportação de roupas por país e gênero. Com base na tabela abaixo, deseja-se identificar o país com a maior proporção de exportações femininas em relação ao total de exportações (feminino + masculino).

Pergunta:
Qual país possui a maior proporção de exportações femininas em relação ao total exportado?

```python
df_grouped = df_confeccoes.groupby(["País de Exportação", "Gênero do Consumidor"])["Receita Total (R$)"].sum().reset_index()
df_grouped.head()
```

```python
# Antes de pivotar: você tem dados em formato longo (linhas com "País", "Gênero", "Receita").
# Após pivotar: você transforma em formato largo, com uma linha por país e uma coluna para cada gênero.
#Isso facilita análises comparativas e cálculos (somar, subtrair, calcular proporção
# resumo:
#Pivotar = reorganizar a tabela para transformar valores de uma coluna (ex: “Gênero”) em colunas separadas, agrupando por outra coluna (ex: “País”

df_pivot = df_grouped.pivot(index="País de Exportação", columns="Gênero do Consumidor", values="Receita Total (R$)").fillna(0)
df_pivot.head()
```

```python
# Agrupa e soma os valores
df_grouped = df_confeccoes.groupby(["País de Exportação", "Gênero do Consumidor"])["Receita Total (R$)"].sum().reset_index()

# Pivot
df_pivot = df_grouped.pivot(index="País de Exportação", columns="Gênero do Consumidor", values="Receita Total (R$)")

# Remove nome do índice da coluna
df_pivot.columns.name = None

# Converte para números
df_pivot["Feminino"] = pd.to_numeric(df_pivot["Feminino"], errors="coerce")
df_pivot["Masculino"] = pd.to_numeric(df_pivot["Masculino"], errors="coerce")

# Calcula a diferença
df_pivot["Diferença"] = df_pivot["Feminino"] - df_pivot["Masculino"]

# Exibe
print(df_pivot.head())
```

```python
# Calcula total exportado e proporção feminina
df_pivot["Total"] = df_pivot["Feminino"] + df_pivot["Masculino"]
df_pivot["Proporção_Feminino"] = df_pivot["Feminino"] / df_pivot["Total"]
df_pivot["Proporção (%)"] = (df_pivot["Proporção_Feminino"] * 100).round(1)

# Exibe a tabela com percentuais
print(df_pivot[["Feminino", "Masculino", "Total", "Proporção (%)"]])

# Descobre o país com maior proporção feminina
pais_maior = df_pivot["Proporção_Feminino"].idxmax()
maior_valor = df_pivot["Proporção_Feminino"].max()

print(f"\nPaís com maior proporção de exportações femininas: {pais_maior} ({maior_valor:.1%})")
```

#### Questão 02 : Cinco empresas de confecção apresentaram seus faturamentos (em reais) por tipo de peça de vestuário. A equipe de análise está interessada em saber qual produto gerou maior faturamento total somando os valores de todas as empresas.

Pergunta:
Qual foi o tipo de peça com maior faturamento total considerando todas as empresas?

```python
# Agrupa por produto e soma o faturamento total
df_product_revenue = df_confeccoes.groupby("Produto")["Receita Total (R$)"].sum().reset_index()

# Encontra o produto com maior faturamento total
produto_maior_faturamento = df_product_revenue.loc[df_product_revenue["Receita Total (R$)"].idxmax()]

print("Faturamento total por produto:")
display(df_product_revenue)

print(f"\nProduto com maior faturamento total: {produto_maior_faturamento['Produto']} (R$ {produto_maior_faturamento['Receita Total (R$)']:.2f})")
```
